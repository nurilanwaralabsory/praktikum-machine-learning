```python
import pandas as pd
from sklearn.model_selection import train_test_split
import numpy as np

# membaca dataset dari file day.csv
df = pd.read_csv('../data/day.csv')
df.head()
```

```python
df_train_full, df_test = train_test_split(
     df, 
     test_size=0.2,
     random_state=42
)

df_train, df_val = train_test_split(
     df_train_full,
     test_size=0.1,
     random_state=42
)

print("--- Data Training ---")
print(f"Jumlah Data: {len(df_train)} baris")
display(df_train.head())
print("\n" + "="*140 + "\n")

print("--- Data Validation ---")
print(f"Jumlah data: {len(df_val)} baris")
print("Lima baris pertama:")
display(df_val.head())
print("\n" + "="*140 + "\n")

print("--- Data Testing ---")
print(f"Jumlah data: {len(df_test)} baris")
print("Lima baris pertama:")
display(df_test.head())
```
